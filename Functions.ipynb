#Q1 :  What is the difference between a function and a method in Python?


ans :   In Python, functions and methods are both blocks of code that perform a specific task. However, there are key differences between them.

# Functions

- Independent blocks of code that perform a specific task.
- Defined with the def keyword.
- Can be called independently by their name.
- Do not belong to a class.

# Methods

- Bound to a class or object.
- Defined inside a class definition.
- Called on an instance of the class.
- Have access to the object's attributes.


```python
# Example


# Function
def greet(name):
    print(f"Hello, {name}!")

# Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, my name is {self.name}!")

# Call function
greet("John")  # Output: Hello, John!

# Create object and call method
person = Person("Jane")
person.greet()  # Output: Hello, my name is Jane!


#In this example, greet is a function that takes a name as an argument, while greet in the Person class is a method that accesses the object's name attribute.
```

#Q6 :  Explain the concept of generators in Python and how they are defined?


ans :  Generators are a type of iterable in Python that can be used to generate a sequence of values instead of computing them all at once and storing them in memory. This makes generators useful for handling large datasets or for creating infinite sequences.

## Defining a Generator

A generator is defined using a function with the yield keyword instead of return. When a generator is called, it returns an iterator that can be used to retrieve the values generated by the generator.


```python
#Here's an example of a simple generator that generates the numbers 0 through 9:

def my_generator():
    for i in range(10):
        yield i

# Create a generator object
gen = my_generator()

# Print the values generated by the generator
for value in gen:
    print(value)
```

## How Generators Work

Here's a step-by-step explanation of how generators work:

1. Generator Function: A generator function is defined using the def keyword, just like a regular function. However, instead of using the return keyword, generator functions use the yield keyword to produce values.
2. Generator Object: When a generator function is called, it returns a generator object. This object is an iterator that can be used to retrieve the values generated by the generator.
3. Yield Statement: When the generator object is iterated over, the generator function executes until it reaches a yield statement. The value specified by the yield statement is returned to the caller, and the generator function's execution is paused.
4. Resuming Execution: When the generator object is iterated over again, the generator function resumes execution from where it left off. This process continues until the generator function finishes executing, at which point the generator object raises a StopIteration exception to indicate that there are no more values to generate.


#Q7 :  What are the advantages of using generators over regular functions?


ans :   Generators offer several advantages over regular functions, including:

## 1. Memory Efficiency
Generators only store the current value being generated, making them memory-efficient for handling large datasets.


```python
#Example:

def generate_numbers(n):
    for i in range(n):
        yield i

# Create a generator that generates 1 million numbers
gen = generate_numbers(10**6)

# Print the first 10 numbers
for _ in range(10):
    print(next(gen))

#In this example, the generator only stores the current number being generated, making it memory-efficient for handling large datasets.

```

## 2. Lazy Evaluation
Generators only compute values as they are needed, which can improve performance by avoiding unnecessary computations.


```python
#Example:

def generate_fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# Create a generator that generates the first 10 Fibonacci numbers
gen = generate_fibonacci(10)

# Print the first 5 numbers
for _ in range(5):
    print(next(gen))

#In this example, the generator only computes the next Fibonacci number when it is needed, making it more efficient than computing all 10 numbers at once.
```

## 3. Flexibility
Generators can be used to create complex sequences of values and can be easily composed together to create new generators.


```python
#Example:

def generate_numbers(n):
    for i in range(n):
        yield i

def generate_squares(n):
    for i in generate_numbers(n):
        yield i ** 2

# Create a generator that generates the squares of the first 10 numbers
gen = generate_squares(10)

# Print the squares
for square in gen:
    print(square)

#In this example, the generate_squares generator is composed from the generate_numbers generator, making it easy to create complex sequences of values.
```

## 4. Improved Performance
Generators can improve performance by avoiding the overhead of creating and returning large lists.


```python
#Example:

import time

def generate_numbers(n):
    numbers = []
    for i in range(n):
        numbers.append(i)
    return numbers

def generate_numbers_generator(n):
    for i in range(n):
        yield i

# Measure the time taken to generate 1 million numbers using a list
start_time = time.time()
numbers = generate_numbers(10**6)
end_time = time.time()
print(f"Time taken using list: {end_time - start_time} seconds")

# Measure the time taken to generate 1 million numbers using a generator
start_time = time.time()
gen = generate_numbers_generator(10**6)
for _ in gen:
    pass
end_time = time.time()
print(f"Time taken using generator: {end_time - start_time} seconds")

#In this example, the generator version is significantly faster than the list version for generating 1 million numbers.
```

#Q10 :  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?


ans : The map(), reduce(), and filter() functions in Python are all part of the functional programming paradigm. Here's a brief overview of each function and how they differ:

##Map()

- The map() function applies a given function to each item of an iterable (such as a list, tuple, or string) and returns a map object.
- It takes two arguments: a function and an iterable.
- The function is applied to each item in the iterable, and the results are returned as a map object.


```python
#Example:

def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(square, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

##Reduce()

- The reduce() function applies a given function to the items of an iterable, going from left to right, so as to reduce the iterable to a single output.
- It takes two arguments: a function and an iterable.
- The function is applied to the first two items in the iterable, then to the result and the next item, and so on, until the iterable is exhausted.


```python
#Example:

from functools import reduce

def add(x, y):
    return x + y

numbers = [1, 2, 3, 4, 5]
sum_of_numbers = reduce(add, numbers)
print(sum_of_numbers)  # Output: 15
```

##Filter()

- The filter() function constructs an iterator from elements of an iterable for which a function returns true.
- It takes two arguments: a function and an iterable.
- The function is applied to each item in the iterable, and items for which the function returns true are included in the resulting filter object.



```python
#Example:

def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(is_even, numbers))
print(even_numbers)  # Output: [2, 4]
```

#Q1 :  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

ans :  Here is a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list:



```python
def sum_even_numbers(numbers):
    """
    Returns the sum of all even numbers in the input list.

    Args:
        numbers (list): A list of numbers.

    Returns:
        int: The sum of all even numbers in the list.
    """
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum


#Here's an example usage of the function:


numbers = [1, 2, 3, 4, 5, 6]
result = sum_even_numbers(numbers)
print(result)  # Output: 12


#Alternatively, you can use the filter() function and a generator expression to achieve the same result in a more concise way:


def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)
```

```python
def reverse_string(input_string):
    """
    Returns the reverse of the input string.

    Args:
        input_string (str): The string to reverse.

    Returns:
        str: The reversed string.
    """
    return input_string[::-1]
```

#Q3 :  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def square_numbers(numbers):
    """
    Returns a new list containing the squares of each number in the input list.

    Args:
        numbers (list): A list of integers.

    Returns:
        list: A new list containing the squares of each number.
    """
    return [num ** 2 for num in numbers]
```

#Q4 :  Write a Python function that checks if a given number is prime or not from 1 to 200.

ans :  Here is a Python function that checks if a given number is prime or not from 1 to 200:


```python
def check_prime(n):
    """
    Checks if a given number is prime or not.

    Args:
        n (int): The number to check.

    Returns:
        bool: True if the number is prime, False otherwise.
    """
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
for i in range(1, 201):
    if check_prime(i):
        print(f"{i} is a prime number")
```

Here's an explanation of how the function works:

- The function takes an integer n as input and returns a boolean value indicating whether n is prime or not.
- If n is less than or equal to 1, it is not prime, so the function returns False.
- The function then checks if n is divisible by any number from 2 to the square root of n. If it is, then n is not prime, so the function returns False.
- If n is not divisible by any of these numbers, then it is prime, so the function returns True.


#Q5:  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


ans : Here is an example of an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms:


```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.i = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.i >= self.n:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.i += 1
        return result

# Example usage:
fib_iterator = FibonacciIterator(10)
for num in fib_iterator:
    print(num)
```

Here's an explanation of how the iterator class works:

- The __init__ method initializes the iterator with the specified number of terms n.
- The __iter__ method returns the iterator object itself, which is required for the iterator protocol.
- The __next__ method returns the next number in the Fibonacci sequence. It uses a simple iterative formula to calculate the next number, and keeps track of the current index i to know when to stop.
- When the iterator is exhausted (i.e., when i reaches n), the __next__ method raises a StopIteration exception to signal the end of the iteration.


#Q6:  Write a generator function in Python that yields the powers of 2 up to a given exponent.


ans : Here is a generator function in Python that yields the powers of 2 up to a given exponent:


```python
def powers_of_two(exponent):
    """
    Yields the powers of 2 up to the given exponent.

    Args:
        exponent (int): The maximum exponent.

    Yields:
        int: The next power of 2.
    """
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power)
```

Here's an explanation of how the generator function works:

- The powers_of_two function takes an exponent argument, which specifies the maximum exponent.
- The function uses a for loop to iterate over the range of exponents from 0 to exponent (inclusive).
- Inside the loop, the yield statement is used to produce the next power of 2. The 2 ** i expression calculates the power of 2 for the current exponent i.
- The yield statement suspends the function's execution and returns the current power of 2 to the caller.
- When the caller requests the next value, the function resumes execution from where it left off and produces the next power of 2.


#Q7:  Implement a generator function that reads a file line by line and yields each line as a string.

ans : Here is a generator function that reads a file line by line and yields each line as a string:



```python
def read_file_line_by_line(file_path):
    """Generator function that reads a file line by line."""
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Stripping newline characters
    except FileNotFoundError:
        print(f"The file {file_path} does not exist.")
```

#Q9 :  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


ans : Here's a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit:


```python
# Define a function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Define a list of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40]

# Use map() to convert the temperatures to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the converted temperatures
print("Celsius Temperatures:", celsius_temperatures)
print("Fahrenheit Temperatures:", fahrenheit_temperatures)

```

- The celsius_temperatures list contains temperatures in Celsius.
- The map() function applies the celsius_to_fahrenheit function to each temperature in the celsius_temperatures list, converting them to Fahrenheit.
- The list() function is used to convert the map object to a list, which is then stored in the fahrenheit_temperatures variable.
- Finally, the program prints the original Celsius temperatures and the converted Fahrenheit temperatures.

#Q10 : Create a Python program that uses `filter()` to remove all the vowels from a given string.

ans : Here's a Python program that uses filter() to remove all the vowels from a given string:



```python
# Define a function to check if a character is a vowel
def is_vowel(char):
    return char.lower() in 'aeiou'

# Define a string
input_string = "Hello, World!"

# Use filter() to remove vowels from the string
vowel_free_string = ''.join(filter(lambda char: not is_vowel(char), input_string))

# Print the original string and the vowel-free string
print("Original String:", input_string)
print("Vowel-Free String:", vowel_free_string)
```

#Q11 : Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:                             
| Order Number | Book Title and Author | Quantity | Price per Item|
|--|--|--|--|
| 34587 | Learning Python, Mark Lutz | 4 |  40.95 |
| 98762 | Programming Python, Mark Lutz | 5 | 56.80 |
| 77226 | Head First Python, Paul Barry | 3 | 32.95 |
| 88112 | Einführung in Python3, Bernd Klein | 3 | 24.99|


 Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map.

```python
# Input data: List of sublists with order details
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99],
]

# Using map and lambda to process the data
result = list(
    map(
        lambda x: (x[0], round(x[2] * x[3] + (10 if x[2] * x[3] < 100 else 0), 2)),
        orders,
    )
)

# Output the result
print(result)
```

Explanation:

1. Input Structure: The orders list contains sublists, where:

x[0] is the order number.

x[2] is the quantity.

x[3] is the price per item.



2. Lambda Function:

Calculates the total price (x[2] * x[3]).

Adds 10 if the total price is less than 100 (+ (10 if x[2] * x[3] < 100 else 0)).



3. map():

Applies the lambda function to each sublist.



4. round():

Rounds the result to 2 decimal places for monetary formatting.
